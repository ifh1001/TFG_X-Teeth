## Manual de Uso

#### Primera Ejecución ( Cada vez que se abre el Notebook)
Para poder usar la aplicación, lo primero que hay que hacer es ejecutar la celda de los Imports para poder cargar todas las funciones necesarias. Para ello es tan sencillo como ponerse en dicha celda y posteriormente dar al botón "Run" que se encuentra en la parte superior. 

Una vez se haya ejecutado (a la izquierda de la celda pone un 1), pasamos a celda de Cargar Imagen. Nos situamos en ella y la ejecutamos de igual manera que la anterior. Una vez esta lista veremos a su izquierda el número 2. También veremos con debajo de esa celda aparecerá un botón que nos permite subir imágenes (Upload).

Si pulsamos en el botón, nos aparecerá una ventana para poder subir la/s imagen/es deseada/s. Una vez las tengamos subidas veremos como el botón tiene en su interior un número, indicándonos que se ha cargado correctamente el número de imágenes deseado.

Ahora vamos a la celda de Calcular Longitud, y tras esperar unos segundos, nos mostrará cada imagen cargada junto con la segmentación, la recta central del diente y una celda con la longitud del diente.

#### Resto de Veces 
Una vez ya hemos realizado la primera ejecución, en caso de que queramos realizar el proceso para más imágenes, basta con pulsar el botón de Upload de nuevo y cargar la siguiente tanda de imágenes. A continuación, no colocamos en la celda de Calcular Longitud y la ejecutamos. Veremos como los resultados son ahora de la nueva tanda. Este proceso se puede hacer tantas veces como se desee.
